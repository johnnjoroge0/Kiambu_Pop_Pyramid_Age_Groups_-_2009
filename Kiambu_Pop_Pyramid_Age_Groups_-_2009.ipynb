```python
#IMPORTING NECESSARY LIBRARIES
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#SELECT DATA SOURCE
path='C:/pydatafiles/Kiambu_Pop_Pyramid_Age_Groups_-_2009.csv'
attributes=["Hours","Marks"]
Data=pd.read_csv(path)
print(Data)
```

```python
#data cleaning
print(Data.isnull().sum())
```

```python
#mesuring the central tendancy
#extract the data
Data_1=Data[["_Age","Female","Male"]]
print(Data_1)
```

```python
#NUMBER OF EXTRACTED COLUMNS
No_of_Female=Data_1["Female"].count()
print("\n No of Female:\n",No_of_Female)
No_of_Male=Data_1["Male"].count()
print("\n No of Male:\n",No_of_Male)
```

```python
#CLCULATING MEAN MEDIAN AND MODE 
Meanfemale=round(Data["Female"].mean(),1)
Meanmale=round(Data["Male"].mean(),1)
print("\n Average Hours\n",Meanfemale)
print("\n Average Hours\n",Meanmale)
Medianfemale=Data_1["Female"].median()
Medianmale=Data_1["Male"].median()
print("\n Median Female\n",Medianfemale)
print("\n Median Marks\n",Medianmale)
Modefemale=Data_1["Female"].mode()
Modemale=Data_1["Male"].mode()
print("\n Mode Female\n",Modefemale)
print("\n Mode Male\n",Modemale)
```

```python
#VISUALIZE MODE USING BAR CHART
import matplotlib.pyplot as plt
Barchart=Data.plot(kind='bar')
plt.show()
```

```python
Data_1["Female"].plot(kind='hist', figsize=(10, 8), linewidth=2, color='yellow', edgecolor='gray')
plt.xlabel("Female Distribution", labelpad=15)
plt.ylabel("Frequency", labelpad=15)
plt.title("Distribution of Female Gender in KIAMBU County", y=1.012, fontsize=22)
measurements = [Modefemale[3],Medianfemale ,Meanfemale]
names = ["mode", "median", "Average"]
colors = ['green', 'blue', 'orange']
for measurement, name, color in zip(measurements, names, colors):
    plt.axvline(x=measurement, linestyle='dotted', linewidth=2.5, label='{0} in {1}' .format(name, measurement), c=color)
plt.legend();
```

```python
# MEASURE OF DISPERSION
#a.i)Calculating the Minimum age
Age_with_min_no_of_female=Data_1[Data_1["Female"]==Data_1["Female"].min()]
print("\n Age with the Least Number of females\n",Age_with_min_no_of_female)
Age_with_min_no_of_male=Data_1[Data_1["Male"]==Data_1["Male"].min()]
print("\n Age with the Least Number of males\n",Age_with_min_no_of_male)

#ii) Calculating the Maximum age
Age_with_max_no_of_female=Data_1[Data_1["Female"]==Data_1["Female"].max()]
print("\n Age with the Largest Number of females\n",Age_with_max_no_of_female)
Age_with_max_no_of_male=Data_1[Data_1["Male"]==Data_1["Male"].max()]
print("\n Age with the Largest Number of males\n",Age_with_max_no_of_male)
```

```python
#B.Calculating the inter Quatile Range
print("Quantile for Males ")
print("--------------------------")
print("Lower Quantile: ",Data_1["Male"].quantile(0.25))
print("Middle Quantile: ",Data_1["Male"].quantile(0.50))
print("Upper Quantile: ",Data_1["Male"].quantile(0.75))
print("Inter Quantile Range: ",Data_1["Male"].quantile(0.75)-Data_1["Male"].quantile(0.25))
```

```python
#CALCULATING THE STANDARD DEVIATION
print("STD for Females",Data_1["Female"].std())
print("STD for males",Data_1["Male"].std())
```

```python
#Calculating variance
print(" Female Variance",(Data_1["Female"].std())**2)
print(" Male Variance",(Data_1["Male"].std())**2)
```

```python
#VISUALIZE MEASURES OF DISPERSION USING BOXPLOT
Data_1["Male"].plot(kind='box', showmeans="true")
plt.show
```

```python
#MEASURE OF RESPONSIBILITY
#CONVARIANCE OF MALES AND FEMALES
Convariance=np.cov(Data_1["Male"],Data_1["Female"])
convariance_matrix=pd.DataFrame(Convariance,index=["Female","Male"],columns=["Female","Male"])
print("covariance of Females and Males:\n",convariance_matrix)
#b. Correlation of Females and Males
corr_females_males=Data_1.corr(method="pearson")
print("\n Correlation of Females and Males\n",corr_females_males)
```

```python
#VISUALIZATION TECHNIQUE USING HEEAT MAP
import seaborn as sns
print("Heat Map to visualize Correlation between Females and Males in Kiambu County")
sns.heatmap(Data_1.corr(),annot=True,fmt=".2g")
```

```python
#VISUALIZATION CORRECTION USING SCATTER GRAPH
from matplotlib import pyplot
pyplot.scatter(Data_1["Age"],Data_1["Male"],color="green")
pyplot.title("Male Population in Kiambu County")
pyplot.xlabel("Age",labelpad=15)
pyplot.ylabel("Male",labelpad=15)
print("Scatter Graph visualize relationship between Age and Gender")
pyplot.show()
```
